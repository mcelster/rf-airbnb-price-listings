<b>Inspecting missing values</b>: 

```python
nan_count = np.sum(df.isnull(), axis = 0)
nan_detected = nan_count != 0
nan_detected
```

```python
##replacing the missing values with the mean only makes sense for the columns that contain numerical values
is_int_or_float = (df.dtypes == np.int64) | (df.dtypes == np.float64)
to_impute = nan_detected & is_int_or_float
to_impute
```

```python
##these columns contain missing values that can be replaced with mean values
to_impute_selected = list(df.columns[to_impute])
to_impute_selected
```

```python
##For every column listed in `to_impute_selected`, fill the missing values with the corresponding mean of all values in the column
for column in to_impute_selected:
    df[column].fillna(value=df[column].mean(), inplace=True)

##Checking results
for colname in to_impute_selected:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))
```

<b>Next, we address outliers using winsorization:</b>

```python
##Create a new column, titled `pricing` by winsorizing the `price` column with the top and bottom 1% percentile values.
import scipy.stats as stats
df['pricing'] = stats.mstats.winsorize(df['price'], limits=[0.1, 0.1])
```

<b>Next, we one-hot encode features that can be one hot encoded:</b>

```python
##One hot encodes the neighborhood group cleansed column so it will show a 1 or 0 for the neighborhood instead of Strings
df_neighborhood_group_cleansed= pd.get_dummies(df['neighbourhood_group_cleansed'], prefix='neighborhood_group_cleansed_')
df_neighborhood_group_cleansed
df.head()
```

```python
##One hot encodes the room type column so it will show a 1 or 0 for the room type instead of Strings
df_room_type= pd.get_dummies(df['room_type'], prefix='room_type_')
df_room_type
```

```python
df= df.join(df_neighborhood_group_cleansed)
df.drop(columns='neighbourhood_group_cleansed',inplace=True)
df= df.join(df_room_type)
df.drop(columns='room_type',inplace=True)
```

```python
##Changing the boolean columns with 0 and 1 so the model can understand
df = df.astype({col: 'int' for col in df.select_dtypes(include=['bool']).columns})
```

```python
#Creating labeled examples
y = df['pricing']
X = df.drop(columns = 'pricing', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1234)
```

```python
from sklearn.model_selection import GridSearchCV

md = [2**n for n in range(2,6)]
msl = [5, 25, 50, 100]
param_grid={'max_depth':md, 'min_samples_leaf':msl}

param_grid
```

```python
print('Running Grid Search...')

# 1. Create a DecisionTreeClassifier model object without supplying arguments

model = RandomForestRegressor(criterion='squared_error', n_estimators=100)

# 2. Run a Grid Search with 5-fold cross-validation using our the model.
#   Pass all relevant parameters to GridSearchCV and assign the output to the object 'grid'

grid = GridSearchCV(model, param_grid, cv=5)

# 3. Fit the model to the training data and assign the fitted model to the 
#    variable grid_search

grid_search = grid.fit(X_train, y_train)

print('Done')
```

```python
best_md = grid_search.best_estimator_.max_depth
best_msl = grid_search.best_estimator_.min_samples_leaf

# Print best performing hyperparameter configuration
print('Best value for max depth: {0}'.format(best_md))
print('Best value for the minimum number of samples required to be at a leaf node: {0}'
      .format(best_msl))

# Print best accuracy score resulting from this configuration
print('Accuracy score: {0}'.format(grid_search.best_score_))
```

<b>Create our optimized model:<b>

```python
print('Begin Random Forest Implementation...')
# 1. Create the RandomForestRegressor model, make sure we are using Regressor and not classifier! Also use criterion 'squared error'
rf_best = RandomForestRegressor(criterion='squared_error', n_estimators=100, max_depth=32, min_samples_leaf=5)

# 2. Fit the model to the training data below
rf_best.fit(X_train, y_train)

# 3. Make predictions on the test data 
rf_preds = rf_best.predict(X_test)
rf_preds

print('End')
```

```python
# The mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, rf_preds)))
# The coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, rf_preds))
```

Below we try implementing a default linear regression model to see if it yeilds better results:

```python
#K fold method to ensure that the rmse & R^2 scores do not change drastically, showing that it can 
# perform well with new data
num_folds = 5
folds = KFold(n_splits = num_folds, random_state=None)

rmse_scores = []
r2_scores = []

for train_row_index , test_row_index in folds.split(X_train): 
    
    # our new partition of X_train and X_val
    X_train_new  = X_train.iloc[train_row_index] 
    X_val = X_train.iloc[test_row_index]
    
    # our new partition of y_train and y_val
    y_train_new = y_train.iloc[train_row_index]
    y_val = y_train.iloc[test_row_index]
    
    rf_best.fit(X_train_new, y_train_new)
    predictions = rf_best.predict(X_val)
    cv = KFold(n_splits=5, shuffle=True, random_state=42)

    rmse = np.sqrt(np.mean((predictions - y_val) ** 2))
    rmse_scores.append(rmse)
    r2 = r2_score(y_val, predictions)
    r2_scores.append(r2)
     
        
for i in range(num_folds):
    print(f'Iteration {i+1} - RMSE: {rmse_scores[i]}, R²: {r2_scores[i]}')
    
print("Mean RMSE:", np.mean(rmse_scores))
print("Mean R²:", np.mean(r2_scores))
```
